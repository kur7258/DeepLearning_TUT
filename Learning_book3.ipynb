```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import mglearn
```

```python
X, y = mglearn.datasets.make_forge()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train, y_train)
```

```python
print(f'Predicting the test set : {clf.predict(X_test)}')
```

```python
print(f'Acurracy of test set : {clf.score(X_test, y_test):.2f}')
```

```python
pig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    #Fit method return self object
    #then, write one-line to create object and fit method
    clf  = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill = True, eps=0.5, ax=ax,
                                   alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title('{} neighbors'.format(n_neighbors))
    ax.set_xlabel('Attribute 0')
    ax.set_ylabel('Attribute 1')

axes[0].legend(loc=3)
```
